```python
# Map text ranges to numeric values (estimated average hours)
sleep_mapping = {
    "Less than 5 hours": 4,
    "5-6 hours": 5.5,
    "6-7 hours": 6.5,
    "7-8 hours": 7.5,
    "More than 8 hours": 9
}
data['Sleep Hours Numeric'] = data['Sleep Duration'].map(sleep_mapping)
```
