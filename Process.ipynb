```python
# Check for missing values
missing_data =  data.isnull().sum()
missing_data
```

```python
# Calculate the percentage of missing values
missing_percentage = data.isnull().sum()/ data.shape[0]
missing_percentage
```

```python
# Fill missing values with the mean for numerical columns

data['Work Pressure'] = data['Work Pressure'].fillna(data['Work Pressure'].mean())
data['Job Satisfaction'] = data['Job Satisfaction'].fillna(data['Job Satisfaction'].mean())

# Check again for missing values
missing_percentage = data.isnull().sum()/data.shape[0] * 100
missing_percentage
```

```python
# Correlation between different factors that potentially lead to depression
import seaborn as sns


# Select variables of interest
cols = ['Work Pressure', 'Job Satisfaction', 
        'Sleep Duration', 'Work/Study Hours', 'Depression']

# Detect categorical variables (object or bool or category types)
cat_vars = [col for col in cols if data[col].dtype == 'object' or data[col].dtype.name == 'category']
num_vars = [col for col in cols if col not in cat_vars]

# Loop through all variable pairs
for i, col1 in enumerate(cols):
    for col2 in cols[i+1:]:
        
        plt.figure(figsize=(6, 4))
        
        if col1 in num_vars and col2 in num_vars:
            # numerical vs numerical → scatter plot
            sns.scatterplot(data=data, x=col1, y=col2)
            plt.title(f"{col1} vs {col2} (Scatter)")

        elif col1 in cat_vars and col2 in num_vars:
            # categorical vs numerical → box plot
            sns.boxplot(data=data, x=col1, y=col2)
            plt.title(f"{col1} vs {col2} (Box)")

        elif col1 in num_vars and col2 in cat_vars:
            # numerical vs categorical → box plot (swap axes)
            sns.boxplot(data=data, x=col2, y=col1)
            plt.title(f"{col1} vs {col2} (Box)")

        else:
            # categorical vs categorical → heatmap of counts
            ctab = pd.crosstab(data[col1], data[col2])
            sns.heatmap(ctab, annot=True, fmt='d', cmap='YlGnBu')
            plt.title(f"{col1} vs {col2} (Count Heatmap)")
        
        plt.tight_layout()
        plt.show()
```

```python
# Compute categorical variables
# Convert Yes/No to 1/0
data['Family History of Depression'] = data['Family History of Depression'].map({'Yes': 1, 'No': 0})
data['Depression'] = data['Depression'].map({'Yes': 1, 'No': 0})
```
