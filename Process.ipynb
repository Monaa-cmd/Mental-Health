# Data Exploration
## * Does deprivation of sleep/job satisfaction/dietary habits lead to higher chances of depression? [T-test/ANNOVA between depressed and group with normal well-being]

```python
# Compute categorical variables, Convert Yes/No to 1/0
data['Depression'] = data['Depression'].map({'Yes': 1, 'No': 0})
```

```python
# Does difference in Work/Study Hours lead to higher chance of Depression?
# Prepare the column for ANNOVA test to see if there are big differences
data['Work/Study Hours'].value_counts()
```

```python
# Create a new categorical column
def categorize_work_hours(hours):
    if hours <4:
        return 'Mild'
    if hours <8:
        return 'Moderate'
    return 'Excessive'
data['Work/Study Hours Category'] = data['Work/Study Hours'].apply(categorize_work_hours)
```

```python
# ANNOVA test
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit ANOVA model
model = ols('Depression ~ C(Work/Study Hours Category)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
