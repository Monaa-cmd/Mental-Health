# Data Cleaning

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\13729\Mental_Health\final_depression_dataset_1.csv")
```

```python
# Check for missing values
missing_data =  data.isnull().sum()
missing_data
```

```python
# Calculate the percentage of missing values
missing_percentage = data.isnull().sum()/ data.shape[0]
missing_percentage
```

```python
# Remove columns with missing values that is higher than 80%
threshold = 0.8
data_cleaned = data.dropna(thresh=threshold * len(data), axis=1)
# Check the shape of the cleaned data
data_cleaned.shape

# Fill missing values with the mean for numerical columns
for column in data_cleaned.select_dtypes(include=['float64', 'int64']).columns:
    data_cleaned[column].fillna(data_cleaned[column].mean())

# Fill missing values with 'Unknown' for categorical columns
# for column in data_cleaned.select_dtypes(include=['object']).columns:
#     data_cleaned[column].fillna(data_cleaned[column].mode()[0], inplace=True)
```
